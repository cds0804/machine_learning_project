```python
import pandas as pd

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_colwidth', -1)

# load file
df = pd.read_csv('naver_webtoon_info.csv')
df.tail()
```

**데이터 전처리**
- 모든 genre에 있는 '스토리' 제거
- 특수기호 제거
- genre 기준 정렬

```python
# genre내부 '스토리' 제거 후 category : content로 진행
# 1. genre에 '스토리' 제거한 폴더를 genre1으로 새로 생성
genre = []
for g in df['genre']:
    genre.append(g.split()[1])
```

```python
df['genre1'] = genre
# content 내부 특수기호 제거
df['content1'] = df['content'].str.replace(pat=r'[%\.\,\?\!\♥\‘\’\'\~\(\)\…\"\<\>\/\+\※\r\n\-\[\]\u200b\']', repl=r'', regex=True)
columns = ['title', 'genre1', 'content1']
df = pd.DataFrame(df, columns=columns)

df
```

```python
# genre1 기준 정렬
df.sort_values('genre1', ascending=True)
```

```python
# genre1을 기준으로 groupby
df1 = df.groupby('genre1').apply(lambda x: x.sum())
df1.T
```
