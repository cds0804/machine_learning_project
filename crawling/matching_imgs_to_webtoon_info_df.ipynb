```python
total_data = []

for dirname in os.listdir('./crawling/naver/imgs/'):
    dirname = unicodedata.normalize('NFC', dirname)
    
    if dirname == '.DS_Store':
        continue
        
    for (dirpath, dirnames, filenames) in walk('./crawling/naver/imgs/{}'.format(dirname)):
        for each_name in filenames:
            copyfile('./crawling/naver/imgs/{}/{}'.format(dirname, each_name), './total_img/{}'.format(dirname + '_' + each_name))
            
            title = list(info_df[info_df['title'] == dirname]['title'])[0]
            genre = list(info_df[info_df['title'] == dirname]['genre'])[0]
            content = list(info_df[info_df['title'] == dirname]['content'])[0]
            age = list(info_df[info_df['title'] == dirname]['age'])[0]
            
            total_data.append({'filename': dirname + '_' + each_name, 
                               'title': title, 
                               'genre': genre, 
                               'content': content, 
                               'age': age})

total_df = pd.DataFrame(total_data, columns=['filename', 'title', 'genre', 'content', 'age'])
total_df.to_csv('total_df.csv')
```
