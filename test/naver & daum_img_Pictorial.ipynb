```python
%%time
import cv2
import os
path = './daum_img/' # 다음 썸네일
file_list = os.listdir(path)
file_list.remove('.DS_Store')
file_list.sort()
len(file_list)
error = []
contour_list_daum = []
for file in file_list:
#     print(file)
    try:
        img = cv2.imread(f'{path}/{file}')
        imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        contour = cv2.Canny(img, 50, 100) # 100, 200은 변경 가능한 인자 값
        contour_list_daum.append(contour)
#         plt.imshow(contour, cmap='gray') # binary
#         plt.xticks([]); plt.yticks([])
#         plt.savefig(f'./contour_naver/{file}', dpi=200)
    except:
        continue
```

```python
contour_daum = np.array(contour_list_daum)
```

```python
np.save('./contour_vector/contour_daum', contour_daum)
```
