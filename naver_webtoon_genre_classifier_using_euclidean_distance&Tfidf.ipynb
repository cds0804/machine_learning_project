```python
# 문장 사이 거리 구하기
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(min_df=1)
```

```python
# 거리를 구하는 것이므로 지도할 내용이 없음
# 훈련용 문장
contents = list(df_naver['content'])
# contents
```

```python
# 형태소 분석
from konlpy.tag import Okt

t = Okt()
def tw_tokenizer(text):
    tokens_ko = t.morphs(text)
    return tokens_ko
```

```python
contents_tokens = [t.morphs(row) for row in contents]
len(contents_tokens)
```

```python
# 형태소 분석 결과를 하나의 문장으로 합침
# 논란의 여지가 있음, 데이터가 많으면 문제가 됨
contents_for_vectorize = []

for content in contents_tokens:
    sentence = ''
    for word in content:
        sentence = sentence + ' ' + word
        
    contents_for_vectorize.append(sentence)
    
# contents_for_vectorize
```

```python
# abs벡터라이즈
X = vectorizer.fit_transform(contents_for_vectorize)
X
```

```python
# 10개의 문장, 전체 말뭉치의 단어 2316개
num_samples, num_features = X.shape
num_samples, num_features
```

```python
# 유클리드 거리
import scipy as sp

def dist_raw(v1, v2):
    delta = v1 - v2
    return sp.linalg.norm(delta.toarray())
```

### 네이버 웹툰 줄거리와 다음 웹툰 줄거리 사이의 거리
#### TF-IDF 인코딩 방법
- Term Frequency - Inverse Document Frequency
- 모든 문서에 공통적으로 들어있는 단어의 경우 문서 구별 능력이 떨어진다고 보아 가중치를 축소하는 방법
- $$\text{tf-idf}(d, t) = \text{tf}(d, t) \cdot \text{idf}(t)$$
- tf(d, t): 특정 단어의 빈도수
- $$\text{idf}(d, t) = \log \dfrac{n}{1 + \text{df}(t)}$$
- idf(t) : 특정 단어가 들어있는 문서의 수에 반비례하는 수
- n : 전체 문서의 수
- df(t) : 단어 t를 가진 문서의 수

```python
# 거리를 구하는 것이므로 지도할 내용이 없음
# 훈련용 문장
contents = list(df_naver['content'])
# contents
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(min_df=0, decode_error='ignore')
```

```python
# 다시 적용
X = vectorizer.fit_transform(contents_for_vectorize)
num_samples, num_features = X.shape
num_samples, num_features
```

```python
# 내용 확인
# X.toarray().transpose()
```

```python
# 테스트 문장 적용
new_post_vec = vectorizer.transform(new_post_for_vectorize)
new_post_vec.toarray()
```

```python
def dist_norm(v1, v2):
    v1_normalized = v1 / sp.linalg.norm(v1.toarray())
    v2_normalized = v2 / sp.linalg.norm(v2.toarray())
    
    delta = v1_normalized - v2_normalized
    
    return sp.linalg.norm(delta.toarray())
```

```python
# 결과
dist = [dist_norm(each, new_post_vec) for each in X]

print("Best post is ", dist.index(min(dist)), ", dist = ", min(dist))
print("test post is : ", new_post)
print("best dist post is : ", contents[dist.index(min(dist))])
```

#### TF-IDF 인코딩
- Term Frequency - Inverse Document Frequency
- 모든 문서에 공통적으로 들어있는 단어의 경우 문서 구별 능력이 떨어진다고 보아 가중치를 축소하는 방법

- tf(d, t): 특정 단어의 빈도수
- $$\text{idf}(d, t) = \log \dfrac{n}{1 + \text{df}(t)}$$
